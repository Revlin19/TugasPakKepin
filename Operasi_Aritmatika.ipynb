````python
# @title Default title text
from typing import cast
#Operasi Aritmatika

#Tambah(+)
a=7
b=12
c=a+b
print(c)
print()

#Kurang(-)
a=80
b=120
c=98
o=a-b-c
print(o)
type(a)
print()

#Perkalian(*)
c=250
d=18*c
print(d)
print()

#Pembagian(/)
c=250
d=18/c
print(d)
print()

#Modulus(%)
a=8500
b=250
c=250
d=18*c
e=a%b%c%d
print(e)
print()

#Exponent(**)
a=8
b=2
e=a**b
print(e)
print()

#Floor Division(//)
a=8
b=2
e=a//b
print(e)
print()

#Addition Shorthand(+=)
a+=2
print(a)
print()

#Subtraction Shorthand(-=)
a-=3
print(a)
print()

#Multiplication Shorthand(*=)
a*=6
print(a)
print()

#Division Shorthand(/=)
a/=8
print(a)
print()

#Equal False(==)
a=8
b=2
e=a==b
print(e)
print()

#Equal True(==)
a=8
b=8
e=a==b
print(e)
print()

#Not Equal False(!=)
a=8
b=8
e=a!=b
print(e)
print()

#Not Equal True(!=)
a=8
b=0
e=a!=b
print(e)
print()

#Not Equal True(!=)
a=8
b=0
e=a!=b
print(e)
print()

#Greater Than(>)
a=8
b=0
e=a>b
print(e)
print()

#Less Than(<)
a=8
b=0
e=a<b
print(e)
print()

#Greater Than or Equal TO(>=)
a=8
b=0
e=a>=b
print(e)
print()

#Less Than or Equal TO(<=)
a=8
b=0
e=a>=b
print(e)
print()

#True if both statement are true(a<5 and b>10)
print(a<5 and b>10)
print()

#True if either statement is true(a<5 or b>10)
print(a<5 or b>10)
print()

#Reverses the result if result is true then false and vice versa(not(a<5))
print NOT(a<5 or b>10)
print()

````
